## 4. Data Adjustment

### 4.1. Analyzing Missing Value Correlation with Ticket Status

```python
df.groupby('Ticket Status')[[
    'Resolution',
    'First Response Time',
    'Time to Resolution',
    'Customer Satisfaction Rating'
]].apply(lambda x: x.isnull().sum())
```

Missing values ​​are assumed to be ongoing processes (not yet closed), so the data is left as is.

### 4.2. Handling Data Types

```python
# Date of Purchase menjadi tipe datetime
df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"],errors="coerce")

# First Response Time menjadi tipe datetime
df["First Response Time"] = pd.to_datetime(df["First Response Time"],errors="coerce")

# Time to Resolution menjadi tipe datetime
df["Time to Resolution"] = pd.to_datetime(df["Time to Resolution"],errors="coerce")
```

## 5. Data Validation

### 5.1. Data Context

```python
# Create a data context
context = FileDataContext.create(project_root_dir='./')
```

### 5.2. Connect to A Datasource

```python
# Datasource
datasource_name = 'csv-customer-support'
datasource = context.sources.add_pandas(datasource_name)

# Data Asset
asset_name = 'customer-support-ticket'
path_to_data = './data/customer_support_tickets.csv'
asset = datasource.add_csv_asset(asset_name,filepath_or_buffer=path_to_data)

# Build Batch Request
batch_request = asset.build_batch_request()
```

### 5.3. Create an Expectation Suite

```python
# Create an expectation suite
expectation_suite_name = 'expectation-customer-support'
context.add_or_update_expectation_suite(expectation_suite_name=expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name=expectation_suite_name
)

# check the validator
validator.head()
```

### 5.4. Great Expectation

#### 5.4.1. Expectation 1: Ticket ID Must Be Unique

```python
validator.expect_column_values_to_be_unique(
    'Ticket ID'
)
```

#### 5.4.2. Expectation 2: Customer Age Must Be Between 0 and 100

```python
validator.expect_column_values_to_be_between(
    'Customer Age',
    min_value=0,
    max_value=100
)
```

#### 5.4.3. Expectation 3: Ticket Status Must Contain Valid Categories

```python
validator.expect_column_values_to_be_in_set(
    'Ticket Status',
    ['Open', 'Closed', 'Pending Customer Response']
)
```

#### 5.4.4. Expectation 4: Customer Age Must Be an Integer

```python
validator.expect_column_values_to_be_in_type_list(
    'Customer Age',
    ['int64', 'int32']
)
```

#### 5.4.5. Expectation 5: Ticket ID Must Not Be Null

```python
validator.expect_column_values_to_not_be_null(
    'Ticket ID'
)
```

#### 5.4.6. Expectation 6: Average Customer Age Must Be Within an Expected Range

```python
validator.expect_column_mean_to_be_between(
    "Customer Age",
    min_value=20,
    max_value=60
)
```

#### 5.4.7. Expectation 7: Date of Purchase Must Follow the YYYY-MM-DD Format

```python
validator.expect_column_values_to_match_strftime_format(
    'Date of Purchase',
    '%Y-%m-%d'
)
```

### 5.5. Save Into Expectation Suite

```python
validator.save_expectation_suite(discard_failed_expectations=False)
```

### 5.6. Checkpoint

```python
# Create a checkpoint
checkpoint = context.add_or_update_checkpoint(
    name="checkpoint-customer-support",
    validator=validator
)

# Run a checkpoint
checkpoint_result = checkpoint.run()
```
